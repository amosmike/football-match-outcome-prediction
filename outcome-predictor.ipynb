```python
matches["date"] = pd.to_datetime(matches["date"])
# matches.dtypes
```

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes # Converts str to categories then categories to number 
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
matches["hour"] = matches["time"].str.replace(":.+","", regex=True).astype("int")
matches["day_code"] = matches["date"].dt.dayofweek
```

```python
# matches["target"] = matches["result"].astype("category").cat.codes
matches["target"] = (matches["result"] == "W").astype("int")
matches
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=500, min_samples_split=10, random_state=1) 
# 50 = number of decision trees, sample_split number of samples per leaf before we split the node, random state means if we run the rf multiple times we'll get same result
train = matches[matches["date"]<'2022-01-01']
test = matches[matches["date"]>'2022-01-01']

predictors = ["venue_code", "opp_code", "hour", "day_code"]
rf.fit(train[predictors], train["target"])
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score # what predition of the time was prediction accurate\
acc = accuracy_score(test["target"], preds)
acc*100
```

```python
# See which situations accuracy was high or low
combined = pd.DataFrame(dict(actual=test["target"], prediction=preds))
# 
pd.crosstab(index=combined["actual"], columns=combined["prediction"]) # 0, 1, 2, draw, loss, win
```

```python
from sklearn.metrics import precision_score
ps = precision_score(test["target"], preds) # precision score doesn't work with non bianary preds/targets
ps
```

```python
# Compute rolling averages for teams
grouped_macthes = matches.groupby("team")
group = grouped_macthes.get_group("Manchester City")
group
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
matches_rolling = matches_rolling.droplevel('team')
matches_rolling.index = range(matches_rolling.shape[0])
matches_rolling
```

```python
def make_predictions(data, predictors):
    train = data[data["date"]<'2022-01-01']
    test = data[data["date"]>'2022-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], prediction=preds), index=test.index)
    ps = precision_score(test["target"], preds)
    return preds, combined, ps
```

```python
pred, combined, precision = make_predictions(matches_rolling, predictors + new_cols)
precision
preds
```

```python
combined = combined.merge(matches_rolling[["date", "team", "opponent", "result"]], left_index=True, right_index=True)
combined
```
